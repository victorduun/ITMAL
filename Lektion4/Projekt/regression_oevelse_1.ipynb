```python
%reset -f

from sklearn import datasets
import matplotlib.pyplot as plt

diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target
f = diabetes.feature_names

fig, axs = plt.subplots(2, 2)
fig.suptitle('Feature X, y scatter plots', fontsize=16)

for i in range(0, 2):
    for j in range(0, 2):
        axs[i, j].scatter(X[:, i*2+j], y)
        axs[i, j].set_title(f[i*2+j])

plt.show()
```

# 1b, 1c, 1d)

1c)

Ser cirka så godt ud som man kunne forvente. Meget svært at forudsige noget præcist udelukkende på baggrund af bmi, men der er en klar opadgående trend.

1d)

Dataene er nogenlunde normalfordelt

```python
%reset -f

# Assignment b:

from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

diabetes = datasets.load_diabetes()
X = diabetes.data
y_true = diabetes.target.reshape(-1,1)
model = LinearRegression()

X_bmi = X[:,2]
#Need to reshape otherwise model.fit complains due to incorrectly shaped array.
X_bmi = X_bmi.reshape(-1,1)

#Fit on BMI (Best correlation - see discussion in Q1a)
model.fit(X_bmi, y_true)

y_pred = model.predict(X_bmi)

print(f"RMSE: {mean_squared_error(y_true, y_pred)**0.5}")



# Assignment c

import matplotlib.pyplot as plt

fig = plt.figure()
plt.title("Raw BMI data + fit")

ax1 = fig.add_subplot(111)

scatter = ax1.scatter(X_bmi,y_true, s=3, color = 'b')
line = ax1.plot(X_bmi, y_pred, color = 'r')
fig.show()

# Assignment d - uses data from b:

y_residual = y_true - y_pred
fig = plt.figure()
plt.title("Residual histogram")
ax1 = fig.add_subplot(111)
ax1.hist(y_residual, density = True, bins=30)

fig.show()
```
